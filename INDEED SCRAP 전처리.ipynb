### 데이터 불러오기
### csv 파일 등 텍스트 파일로 부터 읽어와 생성
- `pd.read_csv(파일경로, sep=',', header, index_col, encoding)`
    - 파일경로 : 읽어올 파일의 경로
    - sep
        - 데이터 구분자. 
        - 기본값: 쉼표

        - 열이름(컬럼이름)으로 사용할 행 지정
        - 기본값: 첫번째 행
        - None 설정: 열 이름으로 순번을(0부터 시작) 사용한다.
    - index_col=정수,컬럼명
        - index 명으로 사용할 열이름이나 순번 지정
        - 생략시 순번을 사용
    - encoding
        - 파일 인코딩
        - 생략시 운영체제 기본 encoding 방식

```python
import nltk
import string
string.punctuation
```

#### NLTK 이용하여 text 토큰화
- **Tokenizer**
    - 문장을 token으로 나눠주는 함수
    - NLTK 주요 tokenizer
        - **sent_tokenize()** : 문장단위로 나눠준다.
        - **word_tokenize()** : 단어단위로 나눠준다.
        - **regexp_tokenize()** : 토큰의 단위를 정규표현식으로 지정
        - 반환타입 : 토큰하나 하나를 원소로 하는 list

```python
title_txt=[]
for text in df['title']:
#     word_tokens = nltk.word_tokenize(text)   # 단어단위로 토큰화단위로 토큰화 (공백 기준)
    reg_tokens = nltk.regexp_tokenize(text, r'\w+') # 토큰단위를 정규표현식으로 지정.
    for token in reg_tokens:
        title_txt.append(token)
print(len(title_txt))
title_txt
```

```python
import collections
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# collections.Counter(): 컨테이너에 동일한 값의 자료가 몇개인지를 파악하는데 사용하는 객체
frequency_d = collections.Counter(title_txt)
# print(frequency_d)

wc = WordCloud(width=1000, height=1000, background_color="black", random_state=0, font_path=r'c:\Windows\Fonts\malgun.ttf')
# width=너비, height=높이, max_words:이미지에 넣을 최대 word 수, mex_font_size:최대 폰트 크기 
# plt.figure(figsize=(10,10))
plt.imshow(wc.generate_from_frequencies(frequency_d))
plt.axis("off")   # 축 생략
plt.show()
```
